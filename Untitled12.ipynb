```python
import os
import numpy as np
import random
import sklearn
from skimage import io
from skimage.transform import resize
import tensorflow as tf
from tensorflow import keras
from keras.models import Model,load_model
import keras.backend as k
import matplotlib.pyplot as plt
from keras.utils import plot_model
from sklearn.datasets import fetch_lfw_pairs
from sklearn.model_selection import train_test_split
from keras.callbacks import ModelCheckpoint,EarlyStopping
from keras.layers import Dense,Conv2D,Flatten,Input,MaxPool2D,Dropout,Lambda,Layer,BatchNormalization,AveragePooling2D,RandomFlip,RandomRotation,RandomZoom,Rescaling
```

```python
dataset = fetch_lfw_pairs(subset = 'train',color = True)
x = dataset['pairs']
y = dataset['target']
```

```python
dataset = []
for i in range(x.shape[0]):
  dataset.append((x[i],y[i]))
```

```python
len(dataset)
```

```python
random.shuffle(dataset)
```

```python
x1_train = []
x2_train = []
y_train = []
for i in range(len(dataset)):
  x,y = dataset[i]
  x1 = x[0]
  x2 = x[1]
  x1_train.append(resize(x1,(128,128)))
  x2_train.append(resize(x2,(128,128)))
  if(y == 0):
    y_train.append(1)
  elif(y == 1):
    y_train.append(0)
```

```python
for i in range(10):
  image1 = x1_train[i]
  image2 = x2_train[i]
  plt.subplot(1,2,1)
  plt.title(y_train[i])
  plt.imshow(image1)
  plt.subplot(1,2,2)
  plt.imshow(image2)
  plt.show()
```

```python
x1_train = np.array(x1_train).astype(np.float32)
x2_train = np.array(x2_train).astype(np.float32)
y_train = np.array(y_train).astype(np.float32)
```

```python
def base_network():
  input = Input((128, 128, 3))
  x = Rescaling(scale=1./255)(input)
  x = RandomFlip("horizontal_and_vertical")(x)
  x = RandomRotation(0.2)(x)
  x = RandomZoom(height_factor=0.2, width_factor=0.2)(x)
  x = BatchNormalization()(x)
  x = Conv2D(32, (3,3), activation="tanh",padding = "same")(x)
  x = AveragePooling2D(pool_size=(2, 2))(x)
  x = Dropout(0.3)(x)
  x = BatchNormalization()(x)
  x = Conv2D(64, (3,3), activation="tanh",padding = "same")(x)
  x = AveragePooling2D(pool_size=(2, 2))(x)
  x = Dropout(0.3)(x)
  x = BatchNormalization()(x)
  x = Conv2D(128, (3,3), activation="tanh",padding = "same")(x)
  x = AveragePooling2D(pool_size=(2, 2))(x)
  x = Dropout(0.3)(x)
  x = BatchNormalization()(x)
  x = Conv2D(256, (3,3), activation="tanh",padding = "same")(x)
  x = AveragePooling2D(pool_size=(2, 2))(x)
  x = Dropout(0.3)(x)
  x = Flatten()(x)

  x = BatchNormalization()(x)
  x = Dense(1024)(x)
  x = Dropout(0.3)(x)
  x = Dense(256, activation="tanh")(x)
  model = Model(input, x)
  return model
```

```python
class Distance(Layer):
  def __init__(self):
    super().__init__()
  def call(self,x,y):
    sum_square = tf.math.reduce_sum(tf.math.square(x - y), axis=1, keepdims=True)
    return tf.math.sqrt(tf.math.maximum(sum_square, tf.keras.backend.epsilon()))
```

```python
def contrastive_loss(y_true, y_pred):
  square_pred = tf.math.square(y_pred)
  margin_square = tf.math.square(tf.math.maximum(1 - (y_pred), 0))
  return tf.math.reduce_mean((1 - y_true) * square_pred + (y_true) * margin_square)
```

```python
base = base_network()
```

```python
input_1 = Input(shape = (128,128,3))
vector_1 = base(input_1)

input_2 = Input(shape = (128,128,3))
vector_2 = base(input_2)
```

```python
distances = Distance()(vector_1,vector_2)
distances = BatchNormalization()(distances)
classifier = Dense(units = 1,activation = "sigmoid")(distances)
model = Model(inputs = [input_1,input_2],outputs = classifier)
```

```python
model = Model(inputs = [input_1,input_2],outputs = classifier)
```

```python
plot_model(model)
```

```python
#checkpoint = ModelCheckpoint('/content/drive/MyDrive/LFW_model.h5',monitor = 'val_loss',save_best_only = True)
```

```python
optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.00001,weight_decay = 0.001)
es = EarlyStopping(monitor = 'val_loss',patience = 20)
model.compile(optimizer = optimizer,loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
history = model.fit(x = [x1_train,x2_train],y = y_train,batch_size = 64,validation_split = 0.1,epochs = 100,callbacks = [es])
```

```python
plt.plot(history.history['loss'],label = "loss")
#plt.plot(history.history['accuracy'],label = "accuracy")
plt.legend()
plt.show()
```
